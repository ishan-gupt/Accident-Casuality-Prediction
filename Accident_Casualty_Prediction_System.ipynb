```python
RoadClass = [ 'Unclassified', 'A', 'B', 'A(M)', 'Motorway', 'C' ]
c1 = [ 0, 0, 0, 0, 0, 0]

with open("Accident_Dataset.csv") as file:
    csvFile = csv.DictReader(file)
    for lines in csvFile:
        if lines['1st Road Class'] == RoadClass[0]:
            c1[0] = c1[0] + 1
            continue
        if lines['1st Road Class'] == RoadClass[1]:
            c1[1] = c1[1] + 1
            continue
        if lines['1st Road Class'] == RoadClass[2]:
            c1[2] = c1[2] + 1
            continue
        if lines['1st Road Class'] == RoadClass[3]:
            c1[3] = c1[3] + 1
            continue
        if lines['1st Road Class'] == RoadClass[4]:
            c1[4] = c1[4] + 1
            continue
        if lines['1st Road Class'] == RoadClass[5]:
            c1[5] = c1[5] + 1
            continue
print(c1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.barh(RoadClass,c1)
plt.xlabel("Count")
plt.ylabel("Road Class")
plt.show()
```
